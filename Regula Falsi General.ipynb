# **MÉTODO DE REGULA FALSI PARA ENCONTRAR UNA RAIZ DE UNA FUNCIÓN GENERAL $f(x)$ POR NÚMERO MÁX DE ITERACIONES Y/O ERROR PORCENTUAL RELATIVO MÁX PERMITIDO.**

Nombre y parámetros de la función:

```python
MetRegFalGen(f,a0,b0,ite,er0):
```
**f** es la función igualada a cero $f(x)=0$

A la cual deseamos encontrarle un cero.

**a0** y **b0** son los límites del intervalo $(a_0,b_0)$ donde tenemos certeza de que la función $f(x)$ tiene una raiz. Tenga en cuenta que estos valores deben cumplir que:
$$f(a_0)\cdot f(b_0)<0$$

**ite** son el número máximo de iteraciones a realizar para aproximarse a la raiz.

**er0** es el error relaitivo porcentual máximo permitido.

*NOTA 1: Recuerde ingresar f por medio de la función de phyton "lambda x: f(x)" y llamando la librería sympy para funciones matemáticas especiales*

*Ejemplo: Calcule un cero de la función $f(x)$, al rededor del intervalo (2.1,2.5) realizando máximo 3 iteraciones con un error relativo porcentual no mayor al 0.01%*
$$f(x)={cos}^{2}(x)+\sqrt{x}-2$$

```python
MetRegFalGen(lambda x: (sp.cos(x))**2+sp.sqrt(x)-2,2.1,2.5,3,0.01)
```
*NOTA 2: Si no conoce el dato de la cantidad de iteraciones a realizar coloque un número grande de iteraciones, por ejemplo 100; si no conoce el error relativo porcentual trabaje un error muy pequeño, por ejemplo por debajo del 0.000001%*

````python
import sympy as sp 
from math import *
  
def MetRegFalGen(f,a0,b0,ite,er0):
 a=a0
 b=b0
 i=1
 x = sp.symbols('x')
 c0=9
 er=101
 #print(f(x))
 while (i<=ite and er >= er0):
   
   ca=c0
   #c0=(a0+b0)/2
   c0=a0-(f(a0)*(b0-a0))/(f(b0)-f(a0))
   c1=f(c0)
   a1=f(a0)
   b1=f(b0)
   er=100*abs(c0-ca)/abs(c0)
   if c1==0:
       i=ite+1
   else:
      if a1*c1>0:
        a0=c0
        i=i+1
       
      else:
          b0=c0
          i=i+1
          
 print("Una raiz de la funcion f(x)=",f(x),"dentro del intervalo (",a,",",b,") es x=",float(c0))
 print("Con un error relativo porcentual de","{0:.4f}".format(float(er)),'%',"y",i-1,"iteraciones")
 
 return 
````

````python
MetRegFalGen(lambda x: x**3+2*x**2+3*x+4,-2,1,10,2)
MetRegFalGen(lambda x: x**2-3,1,2,100,0.05)
MetRegFalGen(lambda x: (sp.cos(x))**2+sp.sqrt(x)-2,2.1,2.5,3,0.01)
MetRegFalGen(lambda x: 5/(4-x)-3,2,3,100,0.000001)
````
